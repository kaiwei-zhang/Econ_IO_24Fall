```python
import pandas as pd
product = pd.read_csv('sample.csv')
product
product.loc[1000]='bacon'
product['ups_match'] = None
product

literature= pd.read_pickle('literature.pkl')

#literature['food_list'] = literature.apply(lambda row: [row['food1'], row['food2']], axis=1)
literature
from pandas_parallel_apply import DataFrameParallel, SeriesParallel

from pandarallel import pandarallel

# Initialize pandarallel
pandarallel.initialize()

def return_value(row,p):
    x = row['product_module_descr']
    y = row['ups_match']
    if (p[0] in x)&(p[1] in x): # find new match
        return p
    elif y==None: # no match, no match yet
        return None 
    return y  #has matched
```

```python
for p in literature['food_list']:
    #print(f"{p[0]},{p[1]}", type(p)) 
    product['ups_match'] = product.parallel_apply(lambda x: return_value(x,p),axis=1)
```
